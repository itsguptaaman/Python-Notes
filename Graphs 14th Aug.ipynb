```python
# To plot 3D graph we need 3 axis x, y and z
x=np.linspace(-1,6,30) # 30 datapoints between -1 to 6
x
```

```python
y=np.linspace(-1,6,30) # 30 datapoints between -1 to 6
y
```

```python
z=x+y
z
```

```python
ax=plt.axes(projection='3d') # for 3D
ax.plot3D(x,y,z)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
```
