```python
# Establishing the connection using sqlalchemy
from sqlalchemy import create_engine
user = 'root'
password = 'admin'
host = 'localhost'
port = 3306
database = 'Aman_db'
mydb=create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")
print("Connection established at",mydb)
```

```python
# Now dumping all the data into mysql database
data.to_sql(con=mydb,name="bank_details",index=False, if_exists="append")
```

```python
import mysql.connector as connection
try:
    db=connection.connect(host="localhost",database="Aman_db",user="root",passwd='admin')
    print(db)
    query="select * from bank_details limit 100"
    cursor=db.cursor()
    cursor.execute(query)
    print(cursor.fetchall())
except Exception as e:
    mydb.close()
    print(e)
```
