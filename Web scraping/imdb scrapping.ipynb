```python
from flask import Flask, render_template, request,jsonify
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
search_string="Sholay"
```

```python
imdb_url="https://www.imdb.com/find?q="+search_string
imdb_url
```

```python
imdb_client=uReq(imdb_url)
imdb_client
```

```python
imdb_page=imdb_client.read()
imdb_page
```

```python
imdb_client.close()
```

```python
imdb_html=bs(imdb_page,"html.parser")
imdb_html
```

```python
big_boxes=imdb_html.findAll("div",{"class":"findSection"})
```

```python
box=big_boxes[0]
```

```python
box.table.tr.td.a['href']
```

```python
movie_url = "https://www.imdb.com" + box.table.tr.td.a['href']
```

```python
imdb_client=uReq(movie_url)
```

```python
movie_page=imdb_client.read()
movie_page
```

```python
imdb_client.close()
```

```python
movie_request=requests.get(movie_url)
movie_request
```

```python
movie_request.encoding="utf-8"
```

```python
movie_html=bs(movie_request.text,"html.parser")
movie_html
```

```python
review_link = movie_url + 'reviews/?ref_=tt_ql_urv'
review_link
```

```python
review_request=requests.get(review_link)
review_request
```

```python
review_request.encoding="utf-8"
```

```python
review_html=bs(review_request.text, "html.parser")
review_html
```

```python
review_boxes = review_html.find_all('div', {'class': 'lister-item-content'})
len(review_boxes)
```

```python
filename = search_string + ".csv"
headers = "MOVIE NAME,REVIEW DATE,REVIEWER,RATING,TITLE,REVIEW \n"
with open(filename,"w") as file:
    file.write(headers)
```

```python
reviews = []
```

```python
for commentbox in review_boxes:
    try:
        # comment_date.encode(encoding='utf-8')
        comment_date=commentbox.find('span',{'class' :'review-date'}).text
    except:
        comment_date='No Comment Date'
    try:
        # name.encode(encoding='utf-8')
        name = commentbox.find('span',{'class':'display-name-link'}).text
    except:
        name = 'No Name'
    try:
        # rating.encode(encoding='utf-8')
        rating = commentbox.div.span.span.text
    except:
        rating = 'No Rating'
    try:
        # comment_head.encode(encoding='utf-8')
        comment_head = commentbox.a.text
    except:
        comment_head = 'No Comment Heading'
    try:
        # long_comment.encode(encoding='utf-8')
        long_comment = commentbox.find('div',{'class':'text show-more__control'}).text
    except:
        long_comment="No long comment available"
```

```python
mydict = {"MOVIE NAME": search_string,"REVIEW DATE": comment_date , "REVIEWER": name, "RATING": rating, "TITLE": comment_head,
                          "REVIEW": long_comment}
```
